```python
import remotestate as rs
```

```python
store = rs.Store({"count": 42, "user": {"name": "Norman"}})
```

```python
class MyService(rs.Service):
    @rs.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)
```

```python
rs.serve(
    MyService(store), 
    ui_dist="./lib-usage",
    mounts={"/dist": "../../remotestate-ts/dist"}, 
    iframe_height=100,
    port=9755,
)
```
