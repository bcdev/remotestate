```python
store = zw.Store({"count": 42, "user": {"name": "Norman"}})
```

```python
class MyService(zw.Service):
    @zw.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)
```

```python
zw.serve(
    MyService(store), 
    ui_dist="./lib-usage",
    mounts={"/dist": "../../remotestate-ts/dist"}, 
    iframe_height=100,
    port=9755,
)
```
