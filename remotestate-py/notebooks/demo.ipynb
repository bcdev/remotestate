```python
import remotestate as rs
```

Assumes you have checked out https://github.com/bcdev/remotestate-demo next to this project
and called `npm run build`. 

```python
store = rs.Store({"count": 42})
```

```python
class MyService(rs.Service):
    @rs.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)
```

```python
rs.serve(
    MyService(store), 
    # ui_dist="http://localhost:5173/",
    ui_dist="../../../remotestate-demo/dist",
    iframe_height=300,
)
```
