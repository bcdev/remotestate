````python
import remotestate as rs
````

````python
store = rs.Store({"count": 42, "user": {"name": "Norman"}})
````

````python
class MyService(rs.Service):
    @rs.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)

    @rs.query
    async def get_count(self) -> int:
        return self.store.get("count")
````

````python
serve_result = rs.serve(
    MyService(store), 
    # Where our "index.html" lives
    ui_dist="./basic-usage", 
    height=100,
    port=9753,
)
````

````python
def on_change(updates):
    print(updates)    
    
store.subscribe(on_change)
````
