```python
import pyre
import logging
```

```python
store = pyre.PythonStore({"count": 42, "user": {"name": "Norman"}})
```

```python
class MyService(pyre.PythonService):
    @pyre.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)

    @pyre.query
    async def get_count(self) -> int:
        return self.store.get("count")
```

```python
pyre.show(service=MyService(store), ui_dist_path="basic-usage", iframe_height=100)
```
