```python
import zwieback as zw
```

Assumes you have checked out https://github.com/forman/zwieback-demo next to this project and called 

```python
store = zw.Store({"count": 42})
```

```python
class MyService(zw.Service):
    @zw.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)
```

```python
zw.serve(
    MyService(store), 
    # ui_dist="http://localhost:5173/",
    ui_dist="../../../zwieback-demo/dist",
    iframe_height=300,
)
```
