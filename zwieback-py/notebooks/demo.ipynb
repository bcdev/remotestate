```python
import zwieback as zw
```

```python
store = zw.Store({"count": 42})
```

```python
class MyService(zw.Service):
    @zw.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)
```

```python
zw.serve(
    MyService(store), 
    dist_dir="../../../zwieback-demo/dist", 
    iframe_height=200,
)
```
