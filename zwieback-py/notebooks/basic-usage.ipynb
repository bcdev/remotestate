```python
import zwieback as zw
```

```python
store = zw.Store({"count": 42, "user": {"name": "Norman"}})
```

```python
class MyService(zw.Service):
    @zw.action
    async def increment(self):
        self.store.set("count", self.store.get("count") + 1)

    @zw.query
    async def get_count(self) -> int:
        return self.store.get("count")
```

```python
zw.serve(MyService(store), ui_dist="./basic-usage", iframe_height=100)
```
